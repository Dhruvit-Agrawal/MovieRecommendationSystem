#Vectorising Dataframe

```python
#Bag Of Words
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
#vectorising 'tag' column
vectors=(cv.fit_transform(df['tags']).toarray())
```

```python
df['tags_vectorised']=list(vectors)
```

```python
df['tags_vectorised'].shape
```

#COSINE SIMILARITY

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_matrix=cosine_similarity(vectors)
```

```python
similarity_matrix #collection of similarity value array of each movie with other 4805 movies
```

```python
similar_movie_list=sorted(list(enumerate(similarity_matrix[0])),key=lambda x:x[1],reverse=True)[1:6]
```

```python
similar_movie_list
```

#RECOMMENDING TOP 5 SIMILAR MOVIES

```python
def recommendMovie(movie):
  #fetch movie index from title
  movie_index=df[df['title']==movie].index[0]
  #fetch similarity scores
  distances=similarity_matrix[movie_index]

  #fetch the indices of the top 5 similar movies
  similar_movie_list=sorted(list(enumerate(distances)),key=lambda x:x[1],reverse=True)[1:6]

  #fetch title of similar movies
  similar_movies_title=[]
  for i in similar_movie_list:
    similar_movies_title.append(df.iloc[i[0]].title)

  return similar_movie_list,similar_movies_title
```

```python
recommendMovie('Avatar')
```
